# Polynomial Linear Regression

## Import Libs and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("polynomial_regression.csv", sep=";")
```

```python
df.head()
```

```python
x = df.araba_fiyat.values.reshape(-1, 1)
y = df.araba_max_hiz.values.reshape(-1, 1) 
```

```python
plt.scatter(x,y)
plt.xlabel("araba_max_hiz")
plt.ylabel("araba_fiyat")
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

## Predict & Visualize

```python
y_head = lr.predict(x)
y_head
```

### Linear Regression is bad decision for this data

```python
plt.scatter(x,y)
plt.ylabel("araba_max_hiz")
plt.xlabel("araba_fiyat")
plt.plot(x, y_head, "red")
```

```python
lr.predict([[10000]])
```

## Formules
### Linear Regression --> y = b0 + ( b1 * x )
### Multiple Linear Regression --> y = b0 + ( b1 * x1 ) + ( b2 * x2 ) + ..... + ( bn * xn )
### Polynomial Linear Regression --> y = b0 + ( b1 * x ) + ( b2 * x^2 ) + .... + ( bn * x^n )

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_reg = PolynomialFeatures(degree=4)
```

```python
x_polynomial = polynomial_reg.fit_transform(x)
```

## Line Fit x_polynomial

```python
linear_reg2 = LinearRegression()
```

```python
linear_reg2.fit(x_polynomial, y)
```

## Predict & Visualize

```python
y_head2 = linear_reg2.predict(x_polynomial)
y_head2[:5]
```

```python
plt.scatter(x,y)
plt.ylabel("araba_max_hiz")
plt.xlabel("araba_fiyat")
plt.plot(x, y_head, "red", label="linear")
plt.plot(x, y_head2, color="green", label="polynomial")
plt.legend()
plt.show()
```
